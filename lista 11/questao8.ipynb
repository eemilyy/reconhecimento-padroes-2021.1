```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread("exercicio-semana11/imagens/Fig0335(a)(ckt_board_saltpep_prob_pt05).png", cv2.IMREAD_GRAYSCALE)
plt.imshow(img, 'gray')
plt.axis("off")
plt.show()
```

filtro de mdia 3  3

```python
ksize = (3, 3)
image = cv2.blur(img, ksize) 

plt.imshow(image, 'gray')
plt.axis("off")
plt.show()
print(image.shape)
```

filtro de mediana 3  3

```python
median = cv2.medianBlur(img, 3)
plt.imshow(median, 'gray')
plt.axis("off")
plt.show()
```
