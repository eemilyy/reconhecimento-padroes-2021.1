```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
```

```python
img = cv2.imread("./exercicio-semana11/imagens/Fig0323(a)(mars_moon_phobos).png", cv2.IMREAD_GRAYSCALE)
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
vals = img.mean(axis=2).flatten()
# plot histogram with 255 bins
counts, bins = np.histogram(vals, range(0, 256))
# plot histogram centered on values 0..255
plt.bar(bins[:-1] - 0.5, counts, width=1, edgecolor='none')
plt.xlim([-0.5, 255.5])
plt.show()
```

```python
img_to_yuv = cv2.cvtColor(img,cv2.COLOR_BGR2YUV)
img_to_yuv[:,:,0] = cv2.equalizeHist(img_to_yuv[:,:,0])
hist_equalization_result = cv2.cvtColor(img_to_yuv, cv2.COLOR_YUV2BGR)
 
cv2.imwrite('result.jpg',hist_equalization_result)


img2 = cv2.imread("result.jpg")
plt.imshow(img2)
plt.axis("off")
plt.show()

vals = img2.mean(axis=2).flatten()
# plot histogram with 255 bins
counts, bins = np.histogram(vals, range(0, 256))
# plot histogram centered on values 0..255
plt.bar(bins[:-1] - 0.5, counts, width=1, edgecolor='none')
plt.xlim([-0.5, 255.5])
plt.show()
```
