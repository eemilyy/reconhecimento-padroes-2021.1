```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#img = cv2.imread("exercicio-semana11/imagens/Fig1038(a)(noisy_fingerprint).png", cv2.IMREAD_GRAYSCALE)
img = cv2.imread("exercicio-semana11/imagens/Fig1039(a)(polymersomes).png", cv2.IMREAD_GRAYSCALE)
plt.imshow(img, 'gray')
plt.axis("off")
plt.show()
```

GLOBAL SIMPLES

```python
t = 170
tNovo = t
delta = 3

img = img.astype(np.uint8)
while True:
    g1 = []
    g2 = []
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if img[i][j] > t:
                g1.append(img[i][j])
            else:
                g2.append(img[i][j])

    mediag1 = np.mean(g1)
    mediag2 = np.mean(g2)

    t = tNovo
    tNovo = (mediag1 + mediag2) / 2

    if (t - tNovo) < delta:
        break

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if img[i][j] > tNovo:
            img[i][j] = 0
        else:
            img[i][j] = 255
    
plt.imshow(img, 'gray')
plt.axis("off")
plt.show()
```
