```python
from sklearn import cluster
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
img = plt.imread("./lirios.jpg")
plt.imshow(img)
```

```python
img.shape
```

```python
print(img)
```

```python
vetor_imagem = np.reshape(img,(375*500,3))
pixels_tabela = pd.DataFrame(vetor_imagem,columns=["R","G","B"])
pixels_tabela
```

```python
def gerar_img(tamanho):
    model = cluster.KMeans(tamanho,init="k-means++",max_iter=40)
    model.fit_predict(vetor_imagem)

    clust_img = model.cluster_centers_[model.labels_] 
    clust_img = np.clip(clust_img.astype('uint8'), 0, 255)
    final = np.reshape(clust_img,(375,500,3))

    plt.imshow(final)
```
